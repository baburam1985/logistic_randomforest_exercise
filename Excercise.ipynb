```python
import pandas as pd


pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', -1)
```

```python
inputDF = pd.read_csv('exercise_01_train.csv')
```

```python
inputDF.head()
```

```python
inputDF = inputDF.convert_objects(convert_numeric=True)
```

```python
inputDF.dropna(inplace=True)
```

```python
inputDF['x41'] = inputDF['x41'].str.lstrip('$')
```

```python
inputDF['x45'] = inputDF['x45'].str.rstrip('%')
```

```python
X = inputDF.iloc[:,:-1]
```

```python
X.head()
```

```python
y = inputDF.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labellec_x = LabelEncoder()
```

```python
X[X['x0'].isnull()]
```

```python
X.iloc[:,34] = labellec_x.fit_transform(X.iloc[:,34])
```

```python
X.iloc[:,35] = labellec_x.fit_transform(X.iloc[:,35])
```

```python
X.iloc[:,68] = labellec_x.fit_transform(X.iloc[:,68])
```

```python
X.iloc[:,93] = labellec_x.fit_transform(X.iloc[:,93])
```

```python
X.x41 = X.x41.astype(float)
```

```python
X.x45 = X.x45.astype(float)
```

```python
onehotencoder = OneHotEncoder(categorical_features=[34,35,68,93],n_values='auto', 
                              handle_unknown='ignore')
```

```python
X = onehotencoder.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=24)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
tempp = classifier.predict_proba(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randClassifer = RandomForestClassifier(n_estimators=20)
```

```python
randClassifer.fit(X_train, y_train)
```

```python
y_pred_rand = randClassifer.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_rand)
```
